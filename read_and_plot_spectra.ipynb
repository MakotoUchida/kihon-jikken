# スペクトルデータを読み込んで描画するスクリプト例

データファアイル`cut_H.png.red(green/blue).csv`を読み込んで、グラフを作成するコード。  
 red,green,blue をそれぞれ `pandas.DataFrame`に読み込んでグラフを作成

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

`pandas.read_csv()`を用いてcsvファイルのデータを読み込む。  
csvファイルは各自アイコンフォルダーをクリックしたのちスペースにドラッグしてアップロードします。

```python
df_r = pd.read_csv('cut_H.png.red.csv',header=None,sep=r'\s+',names=['ch','n'])
df_g = pd.read_csv('cut_H.png.green.csv',header=None,sep=r'\s+',names=['ch','n'])
df_b = pd.read_csv('cut_H.png.blue.csv',header=None,sep=r'\s+',names=['ch','n'])
```

ここで`sep=r'\s+'`はcsvファイルの区切り文字が空白であることを指定するオプションです。  
デフォルトでは`,`が区切り文字となります。

読み込まれたデータの確認

```python
df_r.head() # 最初の５行を表示
# df_g.tail() # 最後の５行を表示
```

# matplotlibで描画する方法

```python
import matplotlib.pyplot as plt
plt.plot(df_r.ch,df_r.n,color="red")
plt.plot(df_g.ch,df_g.n,color="green")
plt.plot(df_b.ch,df_b.n,color="blue")
plt.xlabel('pixel number')
plt.ylabel('count')
```

例えば`(500, 800)` チャンネルを表示したい場合は次のようにします。

```python
plt.plot(df_r.ch,df_r.n,color="red")
plt.plot(df_g.ch,df_g.n,color="green")
plt.plot(df_b.ch,df_b.n,color="blue")
plt.xlabel('pixel number')
plt.ylabel('count')
plt.xlim(500,800) # 範囲指定
```
